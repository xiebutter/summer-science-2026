# Tokenizing Text

LLMs cannot operate directly on strings. Instead, they first convert text into a sequence of integer IDs called tokens. One of the most common tokenization algorithms is Byte Pair Encoding (BPE), which is used by models such as GPT-2.

We first need a numerical representation of text. One approach is to encode every string using UTF-8, which converts each character into one or more bytes. Since a byte can represent values from 0 to 255, every piece of text can be represented as a sequence of integers.

```python
list("Hello world!".encode("utf-8"))
```

Although UTF-8 gives us a numerical representation of text, using raw bytes directly as tokens is inefficient. A tokenizer based only on UTF-8 would have a vocabulary of only 256 possible tokens and very long sequences, since every character (and many Unicode characters) require multiple bytes.

For transformers, long sequences are inefficient because self-attention scales with sequence length. Models also have a limited context window, and longer sequences mean fewer words can fit into the model's context.

Thus, we'd like to represent frequently occurring pieces of text (such as "the" or "ing") with a single token instead of multiple bytes. This is exactly what **Byte Pair Encoding (BPE)** does.

## Byte Pair Encoding (BPE)

Byte Pair Encoding is a simple compression algorithm used for tokenization. The algorithm begins with the original UTF-8 byte vocabulary (256 tokens). It then repeatedly performs the following steps: 

1. Count every pair of consecutive tokens. 
2. Find the most frequently occurring pair.
3. Create a new token representing that pair. 
4. Replace every occurrence of that pair with the new token. 
5. Repeat until the desired vocabulary size is reached. 

Each merge makes the token sequences shorter while increasing the vocabulary size. 

### Load some example text

We'll use a passage containing both ASCII and Unicode characters so we can see how UTF-8 represents different types of text.

```python
# text from https://www.reedbeta.com/blog/programmers-intro-to-unicode/
text = """A Programmer’s Introduction to Unicode March 3, 2017 · Coding · 22 Comments  Ｕｎｉｃｏｄｅ! 🅤🅝🅘🅒🅞🅓🅔‽ 🇺\u200c🇳\u200c🇮\u200c🇨\u200c🇴\u200c🇩\u200c🇪! 😄 The very name strikes fear and awe into the hearts of programmers worldwide. We all know we ought to “support Unicode” in our software (whatever that means—like using wchar_t for all the strings, right?). But Unicode can be abstruse, and diving into the thousand-page Unicode Standard plus its dozens of supplementary annexes, reports, and notes can be more than a little intimidating. I don’t blame programmers for still finding the whole thing mysterious, even 30 years after Unicode’s inception."""
tokens = text.encode("utf-8") # raw bytes
tokens = list(map(int, tokens)) # convert to a list of integers in range 0-255 for convenience
```

Let's inspect the text, the encoded bytes, and their lengths.

```python
print(text)
print("length:", len(text))
print('---')
print(tokens)
print("length:", len(tokens))
```

### Find the most common token pair

The first step of BPE is finding the pair of consecutive tokens that appears most frequently. The function below counts every adjacent token pair.

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]): 
        counts[pair] = counts.get(pair, 0) + 1
    return counts
```

Running the function returns a dictionary where:
* the **key** is a pair of consecutive token IDs
* the **value** is the number of times that pair appears

```python
stats = get_stats(tokens)
print(stats)
```

To better understand the results, we can sort the pairs from most frequent to least frequent.

```python
print(sorted(((v,k) for k,v in stats.items()), reverse=True))
```

The most common pair is the first one in the list, which in this case is (101, 32). We can convert the byte values back into characters to see which actual character pair appears most often.

```python
chr(101), chr(32)
```

### Merge the most common pair

We can identify the most common pair by using Python's `max()` function.

```python
top_pair = max(stats, key=stats.get) # ranking by the value and getting the maximum key
top_pair
```

We now need to create a new token to represent the pair. Since the original byte vocabulary has IDs in the range 0-255, the first new token will receive ID 256. Every occurrence of the selected pair will be replaced by this new token.

The function below scans through the token sequence and replaces every occurrence of the selected pair with the new token ID.

```python
def merge(ids, pair, idx):
    # in the list of ints (ids), replace all consecutive occurences of pair with the new token idx
    newids = []
    i = 0
    while i < len(ids):
        # if we are not at the very last position AND the pair matches, replace it
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            newids.append(idx)
            i += 2
        else:
            newids.append(ids[i])
            i += 1
    return newids
```

Let's test it on a small example.

```python
print(merge([5, 6, 6, 7, 9, 1], (6, 7), 99))
```

Now, we can apply the merge to our text.

```python
tokens2 = merge(tokens, top_pair, 256)
print(tokens2)
```

Notice that the new token sequence is shorter than the original one.

```python
print("length:", len(tokens2))
```

## Decoding: Tokens --> Text

Suppose we already have a sequence of token IDs. To recover the original text, we we first need to create a **vocabulary** that maps each token ID to the sequence of bytes it represents. For the original vocabulary, tokens 0–255 correspond directly to individual UTF-8 bytes.

```python
# a mapping (dictionary) from the token id to the bytes object for that token
vocab = {idx: bytes([idx]) for idx in range(256)}
# go through the tree
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]
```

Once the vocabulary has been reconstructed, the decoding becomes straightforward.

```python
def decode(ids):
    # given ids (list of integers), return Python string
    tokens = b"".join(vocab[idx] for idx in ids) # concatenate bytes together
    text = tokens.decode("utf-8", errors="replace") # not all tokens are standard utf-8, so use errors="replace"
    return text
```

```python
print(decode([90]))
```

## Encoding: Text --> Tokens

Encoding performs the opposite operation. Instead of starting with token IDs, we start with a Python string and want to convert it into the token sequence that the tokenizer has learned. 

We first need to encode the text as UTF-8 bytes. Then we repeatedly apply the BPE merges until no more merges are possible.

Recall that the merges dictionary keeps track of every merge performed during training.

```python
merges
```

The order of these merges is important, because earlier merges create new tokens that later merges depend on.

For example, token **256** was created first. The last merge then combined token **259** with token **256** to create token **275**. Since token **275** depends on token **256**, we need to always apply the earlier merge first.

The encoding algorithm follows these steps:
1. Convert the input text into UTF-8 bytes.
2. Compute all adjacent token pairs.
3. Find the highest-priority merge that exists in the learned vocabulary.
4. Replace that pair with its merged token.
5. Repeat until no more learned merges can be applied.

```python
def encode(text):
    tokens = list(text.encode("utf-8"))
    while len(tokens) >= 2:
        stats = get_stats(tokens)
        pair = min(stats, key=lambda p: merges.get(p, float("inf")))
        if pair not in merges:
            break # nothing else can be merged
        idx = merges[pair]
        tokens = merge(tokens, pair, idx)
    return tokens
```

Let's test the encoder.

```python
print(encode("Hello world!"))
```

Notice that the output is now a sequence of token IDs rather than raw bytes.

## Putting it all together

At this point, we've implemented all of the individual pieces. Let's now organize everything into a single `BasicTokenizer` class. This class relies on two previously implemented helper functions:
1. `get_stats()`
2. `merge()`

```python
class BasicTokenizer():
    def __init__(self):
        self.merges = {}
        self.vocab = {}

    def train(self, text, vocab_size):
        assert vocab_size >= 256
        num_merges = vocab_size - 256
        ids = list(text.encode("utf-8"))
        for i in range(num_merges):
            stats = get_stats(ids)
            pair = max(stats, key=stats.get)
            idx = 256 + i
            print(f"merging {pair} into a new token {idx}")
            ids = merge(ids, pair, idx)
            self.merges[pair] = idx
        self._build_vocab()

    def encode(self, text):
        tokens = list(text.encode("utf-8"))
        while len(tokens) >= 2:
            stats = get_stats(tokens)
            pair = min(stats, key=lambda p: self.merges.get(p, float("inf")))
            if pair not in self.merges:
                break
            idx = self.merges[pair]
            tokens = merge(tokens, pair, idx)
        return tokens
    
    def decode(self, ids):
        tokens = b"".join(self.vocab[idx] for idx in ids) # concatenate bytes together
        text = tokens.decode("utf-8", errors="replace") # not all tokens are standard utf-8, so use errors="replace"
        return text
    
    def _build_vocab(self):
        self.vocab = {idx: bytes([idx]) for idx in range(256)}
        for (p0, p1), idx in self.merges.items():
            self.vocab[idx] = self.vocab[p0] + self.vocab[p1]
```

Once trained, the tokenizer can consistently convert text to tokens and back again, which are the operations needed before an LLM can process any input.
